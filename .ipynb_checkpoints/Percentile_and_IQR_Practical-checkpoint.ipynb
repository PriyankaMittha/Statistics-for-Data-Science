Percentile

A percentile is a measure that indicates the value below which a given percentage of observations in a group of observations fall.

```python
# import all needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Titanic-Dataset.csv")
dataset.head(3)
```

```python
dataset["Age"].isnull().sum()
```

```python
#dataset["Age"].fillna(dataset["Age"].mean(), inplace=True)
```

```python
np.percentile(dataset["Age"],25), np.percentile(dataset["Age"],75)
```

```python
np.percentile(dataset["Age"],50)
```

```python
dataset.describe()
```

```python
# see the outliers using boxplot for Age feature
sns.boxplot(x="Age", data=dataset)
plt.show()
```

```python
# see the outliers using boxplot for Fare feature
sns.boxplot(x="Fare", data=dataset)
plt.show()
```
