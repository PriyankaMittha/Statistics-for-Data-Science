Mean Absolute Deviation

MAD is the average distance between each data point and mean.

```python
#import all libraries are needed
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sec_a and sec_b are two datasets
sec_a = np.array([75,65,73,68,72,67])
sec_b = np.array([90,47,43,51,93,96])
no = np.array([1,2,3,4,5,6])
```

```python
# findout the mean
mean = np.mean(sec_a)
```

```python
#visualize the graph
plt.figure(figsize=(8,3))
plt.scatter(sec_a, no, c="yellow", label="Sec A")
plt.scatter(sec_b, no, c="red",label="Sec B")
plt.plot([70,70,70,70,70,70], no, c="blue", label="Mean")
plt.show()
```

Everytime we cant draw the digram, so we use MAD  to findout the average distance between each data point and the mean

```python
# MAD calculation
mad_a = np.sum(np.abs((sec_a-mean))/len(sec_a))
mad_b = np.sum(np.abs((sec_b-mean))/len(sec_b))
```
